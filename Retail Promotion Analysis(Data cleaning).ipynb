```python
#outliers
#IQR=Q3-Q1
```

```python
numeric_cols = transactions_data.select_dtypes(include=['int64', 'float64']).columns
summary = []

for col in numeric_cols:
    if transactions_data[col].nunique() <= 5 or col=='items_purchased':
        continue
    Q1 = transactions_data[col].quantile(0.25)
    Q2 = transactions_data[col].quantile(0.50)
    Q3 = transactions_data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = transactions_data[(transactions_data[col] < lower) |
                                 (transactions_data[col] > upper)]
    summary.append({
            "Variable": col,
            "Q1": round(Q1, 2),
            "Q2": round(Q1, 2),
            "Q3": round(Q3, 2),
            "IQR": round(IQR, 2),
            "Lower Bound": round(lower, 2),
            "Upper Bound": round(upper, 2),
            "Outlier Count": len(outliers)
        })


    print(col)
    print(f"Lower bound: {lower:.2f}, Upper bound: {upper:.2f}")
    print("Outlier_Count: ", len(outliers))
    print('-'*20,"\n")

#BOX PlOt    
    plt.figure(figsize=(4, 2))
    sns.boxplot(x=transactions_data[col])
    plt.title("Boxplot of " + col)
    plt.show()

#Table of Outliers
summary_df = pd.DataFrame(summary)
summary_df
```

```python
#not all outliers are "bad data" like in COGS and revenue, these represent legitimate large transactions rather than data-entry errors
```
